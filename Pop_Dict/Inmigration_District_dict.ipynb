```python
import pandas as pd
import os
import json
import datetime
import calendar
```

```python
def get_dict(f,dic):
    f="\\"+f
    df=pd.read_excel(r"C:\Users\suhaa\Desktop\Paper\Pop_Dict\Files"+f)
    df=df.loc[(df['Unnamed: 5'] == "All ages") & (df["Unnamed: 4"]=="Total")& (df["Unnamed: 6"]=="Total")]
    for index, row in df.iterrows():
        dic[row["Unnamed: 3"].upper()]=row["Unnamed: 7"]
```

```python
dic={}
i=1
directory=os.path.join(r"C:\Users\suhaa\Desktop\Paper\Pop_Dict\Files")
for root,dirs,files in os.walk(directory):
    for file in files:
        if file.endswith(".XLSX"):
            get_dict(file,dic)
```
