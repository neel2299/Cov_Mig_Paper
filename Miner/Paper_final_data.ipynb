```python
import pandas as pd
df=pd.read_csv("INDIVIDUAL_RUNS.csv")
df
```

```python
destinationlis=[]
originlis=[]
for index, row in df.iterrows():
    try:
        string=row[21]
        lis=string.split(",")
        origin = lis[1].strip(",").strip("']").strip()
        destination = lis[-1].strip(",").strip("']").strip()
        originlis.append(origin)
        destinationlis.append(destination)
    except:
        print(lis)
```

```python
df["origin"]=originlis
df["destination"]=destinationlis
df
```

```python
df.to_csv("Orig-Des_Data.csv")
```
