```python
import pandas as pd
df=pd.read_excel("INDIA.XLSX")
df=df.loc[(df["Unnamed: 4"]=="Total") & (df["Unnamed: 5"]=="All ages" )& (df["Unnamed: 6"]=="Total")]
df
```

```python
dic={}
for index, row in df.iterrows():
    dic[row["Unnamed: 3"].lstrip("State -")[0:-5].upper()]=row["Unnamed: 7"]
dic.pop("")
dic
```

```python
lis=[]
for key in dic.keys():
    lis.append([key,dic[key]])
out=pd.DataFrame(lis)
out.to_csv("EMPLOYMENT-UN.csv")
```
